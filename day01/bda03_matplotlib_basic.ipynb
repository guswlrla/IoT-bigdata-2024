##### 맷플롯립은 한글이 깨짐
- 나눔폰트 옵션 사용
- 폰트관리자 설정

````python
# matplotlib 사용(시 아래내용 설정 필수)
import warnings
import matplotlib.pyplot as plt # plt로 줄여씀
from matplotlib import rcParams, font_manager, rc 

# 나눔고딕코딩 폰트 경로 
font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf'
font = font_manager.FontProperties(fname=font_path, size=10).get_name()

# warning 표시 끄기
warnings.simplefilter('ignore') # 경고메세지 무시 

# 기본 폰트 설정
plt.rcParams['font.family']= font
# 기본폰트 사이즈 설정 
plt.rcParams['font.size'] = 14
# 기본차트 사이즈 변경
plt.rcParams['figure.figsize'] = (8, 6)
# 기본 그리드 표시
plt.rcParams['axes.grid'] = True
# 한글화 후 - 표시가 네모로 나오는 현상 수정 
plt.rcParams['axes.unicode_minus'] = False
````

````python
## 스캐터차트
x = np.random.randint(-100, 100, 50) # -100 ~ 100 사이에 50개만 랜덤
y = np.random.randint(-200, 200, 50)
size = np.random.randint(-10, 200, 50)
color = np.random.randint(0, 100, 50)

plt.scatter(x, y, s=size, c=color, alpha=0.7)
plt.title('온도(화씨)')
plt.colorbar()
plt.show()
````

````python
## 파이차트
size = [2441, 2312, 1031, 1233]
label = ['A','B','AB','O']
color = ['magenta', 'deeppink', 'hotpink', 'pink']

plt.axis('equal') # 가로축, 세로축 스케일 동일
plt.pie(size, labels=label, autopct='%.1f%%', explode=(0,0,0.1,0), colors=color)
plt.legend(loc='lower left')
plt.title('혈액형별 인구율')
plt.show()
````
